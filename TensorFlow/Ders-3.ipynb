```python
df = df.drop("transmission",axis=1)
```

```python
y = df['price'].values
x = df.drop('price',axis=1).values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test= scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential() # modelimizi oluşturuyoruz.

model.add(Dense(12,activation = "relu")) # katman ekleme 
model.add(Dense(12,activation = "relu")) 
model.add(Dense(12,activation = "relu")) 
model.add(Dense(12,activation = "relu")) 

model.add(Dense(1)) # Çıkış katmanıdır. Aktivasyon fonksiyonu koymaya gerek yok

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,epochs=300,validation_data=(x_test,y_test),batch_size=250)
```

```python
kayıpveri = pd.DataFrame(model.history.history)
```

```python
plt.plot(kayıpveri)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
tahmin = model.predict(x_test)
```

```python
tahmin
```

```python
mean_absolute_error(y_test,tahmin)
```

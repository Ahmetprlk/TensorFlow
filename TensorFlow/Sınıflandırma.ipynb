```python
y =df['Type'].values
```

```python
x = df.drop('Type',axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation ,Dropout
from tensorflow.keras.callbacks import  EarlyStopping 
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=10,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) # çıkış katmanınasigmoid fonksiyonu verdik.
 

model.compile(optimizer="adam",loss= "binary_crossentropy")
```

```python
model.fit(x=x_train,y=y_train,epochs=500,validation_data=(x_test,y_test),verbose=1)
```

```python
model.history.history
```

```python
model2 = pd.DataFrame(model.history.history)
```

```python
model2.plot() 
# validation loss ile loss arasındaki fark azaldığında epochs ayarla. Fark fazla olmamalı val_loss ile loss arasında
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=10,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) # çıkış katmanınasigmoid fonksiyonu verdik.
 

model.compile(optimizer="adam",loss= "binary_crossentropy")
```

```python
earlyStopping =EarlyStopping(monitor="val_loss",verbose=1,mode="min",patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=500,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
#EarlyStopping sayesinde epochs erkenden durduruldu
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=10,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation="sigmoid")) # çıkış katmanınasigmoid fonksiyonu verdik.
 

model.compile(optimizer="adam",loss= "binary_crossentropy")
```

```python
model.fit(x=x_train,y=y_train,epochs=500,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```
